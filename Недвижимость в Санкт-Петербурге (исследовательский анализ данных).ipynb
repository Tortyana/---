### Предобработка данных

```python
#рассмотрим, какие значения стоят за столбцами с пропусками
display(data['ceiling_height'].value_counts())
display(data['floors_total'].value_counts())
display(data['living_area'].value_counts())
display(data['is_apartment'].value_counts())
display(data['kitchen_area'].value_counts())
display(data['balcony'].value_counts())
display(data['locality_name'].value_counts())
display(data['airports_nearest'].value_counts())
display(data['cityCenters_nearest'].value_counts())
display(data['parks_around3000'].value_counts())
display(data['parks_nearest'].value_counts())
display(data['days_exposition'].value_counts())
data['ponds_around3000'].value_counts()
```

Заполним пропуски.

```python
#пропусков в высоте потолков более 10%, поэтому в избежании искажения данных их заполнять не будем

#высоту здания возьмем среднюю для конкретного населенного пункта,
#при этом если она ниже конкретного этажа, то берем значение этажа
# создадим вспомогательный список уникальных locality_name
locality_name_list = list(data['locality_name'].dropna().unique())

# создадим вспомогательную таблицу со средней этажностью домомв в населенных пунктах
mean_floors_total_locality = data.pivot_table(index='locality_name', values='floors_total', aggfunc='mean')
mean_floors_total_locality['floors_total'] = mean_floors_total_locality['floors_total'].astype(int)

# с помощью цикла присваиваем Nan среднюю для конкретного населенного пункта этажность
for i in locality_name_list:
    data.loc[(data['floors_total'].isna()) \
                         & (data['locality_name'] == i), 'floors_total'] = mean_floors_total_locality['floors_total'][mean_floors_total_locality.index == i].median()

#для строчек с отстутствующим названием населенного пункта число этажей в доме floors_total заполним медианным значением
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())    
    
#запустим функцию для случаев превышения фактического этажа жилья этажности дома    
def func_floors_comparison(row):
    if row['floor'] > row['floors_total']:
        return row['floor']
    else:
        return row['floors_total']        
        
data['floors_total'] = data[['floors_total', 'floor']].apply(func_floors_comparison, axis=1)


#отработаем пропуски в living_area. Вероятно пользователи сервиса при заполнении поленились заполнить информацию, которую не сочли важной. Т.к. есть информация об общей площади и кухне
#можем посчитать общий % жилой площади от общей площади, и использовать его при заполнении пустых значений
mean_proportion_living_area = (data['living_area'] / data['total_area']).mean()
mean_proportion_living_area
# заполним пустые ячейки методом fillna 
data['living_area'] = data['living_area'].fillna(data['total_area']*mean_proportion_living_area)

# для данного поля целесообразно принять пропуски как False
data['is_apartment'] = data['is_apartment'].fillna(False)


# Nan в kitchen_area более 10%, поэтому в избежании искажения данных их заполнять не будем

#число балконов balcony заполним нулями
data['balcony'] = data['balcony'].fillna(0)

#в параметре locality_name в строчки с пропусками добавим название 'Другой'
data['locality_name'] = data['locality_name'].fillna('Другой')
data['locality_name'].isnull().sum() 

median_airports_nearest_locality = data.pivot_table(index='locality_name', values='airports_nearest', aggfunc='median')
median_airports_nearest_locality.head()

# заменим значения циклом 
# не берем Санкт-Петербург т.к. в нем брать медианные значения будет слишком грубо
# по остальным пропускам отработаем ниже
for i in list(median_airports_nearest_locality.index):
    if i not in ['Санкт-Петербург', 'unknown']:
        data.loc[(data['locality_name'] == i)\
                       & (data['airports_nearest'].isna()), 'airports_nearest'] =\
         float(median_airports_nearest_locality[median_airports_nearest_locality.index==i]['airports_nearest'])
        
# заменим значения циклом  на общую медиану (кроме Питера)
for i in list(median_airports_nearest_locality.index):
    if i not in ['Санкт-Петербург']:
        data.loc[(data['locality_name'] == i)\
               & (data['airports_nearest'].isna()), 'airports_nearest'] =\
        data['airports_nearest'].median()
        
        
#Не заполняем пропуски airports_nearest в Санкт-Петербурге т.к. в нем брать медианные значения будет слишком грубо. Пока оставляем эти пропуски как есть.

#отработаем пропуски в cityCenters_nearest, заменим их на медианное значение в каждой локации
# подготовим вспомогательную таблицу
median_cityCenters_nearest_locality = data.pivot_table(index='locality_name', values='cityCenters_nearest', aggfunc='median')
median_cityCenters_nearest_locality.head()

# заменим значения циклом 
for i in list(median_cityCenters_nearest_locality.index):
    if i != 'unknown':
        data.loc[data['cityCenters_nearest'].isna() &(data['locality_name']==i), 'cityCenters_nearest'] = float(median_cityCenters_nearest_locality[median_cityCenters_nearest_locality.index==i]['cityCenters_nearest'])
        
#отработаем пропуски в parks_around3000 и ponds_around3000.
# приравниваем Nan к 0
data['parks_around3000'] = data['parks_around3000'].fillna(0)        
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)   
   
# в'days_exposition' пропуск может означать, что объявление еще не снято с публикации,
# поэтому заполнять пропуски в days_exposition нулем неправильно      
    
#проверяем количество пропусков
data.isna().sum()
 
```
